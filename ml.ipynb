```python
df = pd.read_csv("https://raw.githubusercontent.com/shsarv/Machine-Learning-Projects/main/Diabetes%20Prediction%20%5BEND%202%20END%5D/dataset/kaggle_diabetes.csv")
```

```python
y=df["Outcome"]
X=df.drop("Outcome",axis=1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.25,random_state = 42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
```

```python
lr = LogisticRegression()
dr = DecisionTreeClassifier()
```

```python
lr.fit(X_train,y_train)
dr.fit(X_train,y_train)
```

```python
print("logistic",lr.score(X_train,y_train),lr.score(X_test,y_test))
print("desciesion tree",dr.score(X_train,y_train),dr.score(X_test,y_test))
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cross_val_score(lr,X_test,y_test,cv=5)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = { 
    'C':[0.1,1,10,1000],
    'penalty': ['l2','l1','none'],
    'solver': ['lbfgs','liblinear']
}
```

```python
grid_search =GridSearchCV(estimator=lr,param_grid=param_grid,cv=5,scoring='accuracy')
```

```python
grid_search.fit(X_train,y_train)
```

```python
best_params = grid_search.best_params_
print("Best Parameters: ",best_params)
grid_search.score(X_test,y_test)
```

```python
print(classification_report(y_test,lr.predict(X_test)))
```
